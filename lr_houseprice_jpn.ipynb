<h2 style="color:blue" align="center"> House Price Prediction- Linear Regression </h2>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import warnings as wr
wr.filterwarnings('ignore')
```

```python
df = pd.read_csv('house-prices.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

<h3 style="color:blue" align="left"> Basic Plots </h3>

#### Histplot

```python
# Price - Frequencey table
df['price'].plot(kind='hist', bins=20, title='Price')
plt.gca().spines[['top', 'right']].set_visible(False)
```

```python
# Area- Freq Table
df['area'].plot(kind='hist', bins=20, title='Area')
plt.gca().spines[['top', 'right']].set_visible(False)
```

#### Scatterplot

```python
plt.scatter(df.area, df.price) # plt.scatter(x, y)
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area vs Price')
plt.show()
```

* By the above visualization, it obvious that, there is linear relationship between Area & Prcie (Positive Relationship)
* So, lets drive the apply Linear Model

<h3 style="color:blue" align="left"> Seperation </h3>

```python
X = df[['area']]
y = df['price']
```

```python
print(X.shape)
print(y.shape)
```

#### Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

<h3 style="color:blue" align="left"> Model </h3>

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

<h3 style="color:blue" align="left"> Prediction </h3>

```python
model.predict([[5000]])
```

<h3 style="color:blue" align="left"> Evaluation </h3>

```python
model.score(X_test, y_test)
```

<h3 style="color:blue" align="left"> Visualization </h3>

```python
plt.scatter(df.area, df.price, color='purple')
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.plot(df.area, model.predict(df[['area']])) # full data ( train & test); area must be in 2D
plt.show()
```

<h3 style="color:blue" align="left"> Prediction Model- Input and Prediction </h3>

```python
# Linear Model
n = input("Enter Area: ")
array = np.array(n)             # the input is converted to numpy array for better computational
array2 = array.astype(np.float32)
value = ([[array2]])            # Input in 2D

result = model.predict(value)   # Predict the value
home_price = np.array(result)  
home_price = home_price.item()  # item() method is used to convert single-element array to a scalar .

print("Home Price: ", home_price)
```
