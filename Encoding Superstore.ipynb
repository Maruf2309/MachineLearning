## Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df1.head()
```

### Label Encoder by Loop

```python
df4.head()
```

```python
# Label Encoder using for loop
import numpy as np

for column in df4.columns:
    if df4[column].dtype == np.number:
        continue
    df4[column] = le.fit_transform(df4[column])
    
df4.head()
```

## Another Way to Encode

```python
# pandas.core.dtypes - is numeric
```

```python
df5.head()
```

```python
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for column in df5.columns:
    if is_numeric_dtype(df5[column]):
        continue
    else:
        df5[column] = le.fit_transform(df5[column])
```
