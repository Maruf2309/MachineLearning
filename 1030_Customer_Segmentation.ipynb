# Clustering

```python
# Model Initialization / Prepare Model

from sklearn.cluster import KMeans
cluster = KMeans()                     # defualt k=8
cluster.fit(df1[['Annual Income (k$)', 'Spending Score (1-100)']])                          # Does not required train test, because lable is unkown
```

```python
# Cluster Default Center values
cluster.cluster_centers_
```

- First pair is center of cluster 1
- 2nd pair is center of cluster 2 & last one is center of cluster 8 (default)
- First column is centroid value for "Annual Income" & 2nd Columns is value for "Spending Score"

- In kmeans clustering, train test not needed, becasue label is unknown, lets say you have been assiged to to a math, but which solution is uknown, so how can be test or trained
- But, its requires validation, to justify how meaningful the result. Silhouttee score is preferable validation technique

```python
cluster.n_clusters, cluster.inertia_       # Cluster must be 8, since we have taken default k
```

# Finding K with Elbow Method

- How can we confirm/validate this above default K=8 is best cluster and 37.73640346254161 this is minimum error
- Defining n_cluster/ k from (2-10) one by one and will try to find out elbow point
- In each cluster we will calculate this intertia & then we will plot it & eventually will try to find elbow points

```python
inertia = np.array([0,0])  # first value is number of K & 2nd value is measures/inner variance intertia for that k value
inertia
```

```python
inertia = np.array([[0,0]])
np.append(inertia,[[1,34]])   # Lets examine by giving k=1 when intertia = 34 ; This is random number for test purpose
```

```python
inertia = np.array([[0,0]])
np.append(inertia,[[1,34]], axis=0)
```

```python
# Complete Process
inertia = np.array([[0,0]])       # like a container, contains k number & interia
for i in range(2,10):
    kcluster = KMeans(n_clusters=i)            # Create k cluster value/ Model
    kcluster.fit(df1[['Annual Income (k$)', 'Spending Score (1-100)']]) # Fitting the Model by its features/ variables
    inertia = np.append(inertia, [[i, kcluster.inertia_]], axis=0)   # inertia + iteration num + cluster intertia , axis as col
    
    
inertia          # return numpy array
```

- As per k values, its found that, at cluster 5, there is a sharp deep (from 108 to 65, then its steady) in wcss/error, so it 5 will be optimal k values
- The more cluster, the less intertia/ iner-variance of cluster
- No matter how many iterations processed, optimal k will be 5 here

### Elbow Visualization

```python
sns.lineplot(x=inertia[:,0], y=inertia[:,1])    # Primary 0 need to drop
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x=inertia[1:,0], y=inertia[1:,1]) 
plt.title("Elbow Visualization")
plt.xlabel('Number of Iteration')
plt.ylabel('Ineria')
plt.show()
```

- (x=inertia[:,0], y=inertia[:,1])  all rows, 1st col that is k value/iteration | 1st col will be 0 & 2nd col will be 1
- (x=inertia[1:,0], y=inertia[1:,1]) : drop original rows number that is 0:0 that why it starts from 1 (ignore top 0;0)
- Now as per pairplot, its proved that number of cluster is 5
- If pairplot/ visualization dont give any clear understanding about optimal number of cluster, ELBOW METHOD is the ultimate/ last resort of finding/ assuming optimal k values

# Visualizing Clusters - By Optimal K Values

```python
# Modeling for Prediction
cluster = KMeans(n_clusters=5)  # Modeling by actual optival k values e.g 5
cluster.fit(df1[['Annual Income (k$)', 'Spending Score (1-100)']]) # fit the model
```

```python
# Prediction
cluster.predict(df1[['Annual Income (k$)', 'Spending Score (1-100)']])  # return cluster name/values ( 5 cluster ) as np array (We can plot by 2 ways either plt or sns)
```

- This prediction cluster name or values we can add to original dataframe for understanding

```python
# Adding Predicted name to original dataframe (e.g df)- For understanding purpose or helpfur for ploting
df['Cluster_Name'] = cluster.predict(df1[['Annual Income (k$)', 'Spending Score (1-100)']])  # Better to do original, not scaled
df.head()
```

```python
# Cluster Visualization as per Optimal K values (as per findings)
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster_Name')
plt.title('Cluster Visualization by Optimal K values')
plt.show()
```

##### Findings:

- We could do the visualization by plt, but that requries one by one, sns helps to plot all together
- K values/ cluster visualization
- If any values not assiged to any cluster, that handled by "DBSCAN CLUSTERTING" algorithm
- DBSCAN: Dont consider any outliers/ noise values, rather is replaced by -1 (In DBSCAN  -1 means = outliers/ noise points)
- Kmeans Clustering: it considers outliers (see the cluster as per color); but DBSCAN do not consider any outliers
- So, Here is 2 approach: if you want to apply kmeans then <b>outliers must be treated/ removed</B> or if you want to go with outliers then go for "DBSCAN CLUSTERING" algorithm. 
- In reallife dataset, sometimes we can treat outliers and sometime its very tough/ impossible to treat/remove these outliers then here we can apply "DBSCAN CLUSTERING ALGORITHM"
- Elbow method we can apply only for Kmeans, Silhouette_Score is global.It can be applied for validation for Kmeans/ Hierarchical & DBSCAN
- Elbow method used for choosing best k values, but to choose the best Cluster model we use <b> Silhouette_Score</b>. That measures how correctly/ properly <b> MODEL </b> can do clustering or can select best numbers of clusters 

- All done, now we need to validate or proof it whether clusting done perfectly - go <b> VALIDATION </b>


# Validation or Model Validation

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(df1[['Annual Income (k$)', 'Spending Score (1-100)']], df['Cluster_Name'])
```

- Silhouette Score- is one of the best approach for clustering validation
- silhouette_score(x, label): it will take x (feastures) and the label that is ('Cluster_Name') as added as new col in df
- silhouette_score : .5545, the score is more closer to 1 consider the more accurate/ better model for clustering, this is not accuary, its a range. Its range (-1 to +1). Negative Score: Poor, Posive Score: Better/Good
- <b> Acceptance range of Silhouette_score: </b> at least .50 or >=.50 is generally recommended
- <b> Silhouette_Scored </b> used for internal validation
- <b> How Silhouette_Score Works : </b> It check & verify the values assigned to clusters (like cluster name) it is ok or not, by calculating a(i) , b(i). This is not Cross-validation, rather this is only validation
- Here, income & spending have been used as features, if need to go for multiple features that also can be defined by Silhouette score, so <b> Silhouette Score </b> helps to find <b> Best Model </b> & <b> Best Features </b> by its <b> Validation Technique </b>. Besides, if muliple model (Kmeans, Hierarchical, DBSCAN) applied, then <b> Silhouette Score </b> help to select best model out of this.
- For Clustering, Kmeans & DBSCAN is very powerful, like in reallife clustering problem around 70-80% problem can be solved by this two clustering algorithms


 <b> Silhouette Coefficient</b> : is calculated using mean intra-cluster distance (a) and the mean nearest-cluster distance (b) for each sample. The Silhouette Coefficient for a sample is (b - a)/max(a,b)
