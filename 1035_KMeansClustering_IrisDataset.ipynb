### Step 1: To start let’s import the following libraries.

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
%matplotlib inline

from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

### Step 2. Load the data

```python
iris = load_iris()
iris
```

### Step 3. Define your target and predictors

```python
df = pd.DataFrame(data=iris.data, columns=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'])
df['target'] = pd.Series(iris.target) # Getting a target col
df['target_names'] = pd.Series(iris.target_names)

species =[]
for i in range(len(df)):
    if df.iloc[i]['target']==0:
        species.append('setosa')
    elif df.iloc[i]['target']==1:
        species.append('versicolor')
    elif df.iloc[i]['target']==2:
        species.append('virginica')
        
df['Species'] = species
```

```python
df.head()
```

```python
x = iris.data # seperate data from df, 4 cols
x
```

### Step 4: Data Visualization through a scatter plot.

```python
plt.figure(figsize=(12,7))
plt.scatter(x=df['sepal length (cm)'], y=df['sepal width (cm)'],c=iris.target,  cmap='gist_rainbow')  # c: Target (setosa, versicolor, virginica)


plt.xlabel('Sepal Length', fontsize=18)
plt.ylabel('Sepal Width', fontsize=18)
plt.show()
```

C = iris.target : Here C is for color as we have 3 values in target, here 3 categories will be differenciated by the color

```python
sns.pairplot(df.drop(['target'], axis=1), hue='Species', height= 2.5, markers=["8","s","D"])
plt.show()
#Observations
#joint distrubution in the triangles
#marginal distribution in the slant line
#'setosa' extremly short petal width
#, 'versicolor', 
#'virginica have high sepal length
# We have 4 features 'sepal length, lepal width, petal length, petal width'
# diagnal plot is called "Marginal distribution or density plot" others plot is called joined distribution
```

### Step 5: Using the elbow method to estimate the value k

```python
#To start, let’s arbitrarily assign the value of k as 5. We will implement k-means clustering using k=5.

kmeans5 = KMeans(n_clusters=5, init='k-means++', random_state=0)
y = kmeans5.fit_predict(x)
print(y)                  # return 0-4,e.g 5 clusters
```

```python
kmeans5.cluster_centers_   # Five Clusters & Centroids for 4 features
```

#### Visualize Datapoints

```python
#x[y==0,0] finds all the rows of x that have a y value of 0 (y == 0), and are in the first column of X i.e sepal length
# Similarly  x[y==0,1] finds all the rows of x that have a y value of 0 (y == 0), and are in the second column of X i.e sepal width

plt.scatter(x[y==0,0], x[y==0,1], s=15, c='red', label='Cluster_1')
plt.scatter(x[y==1,0], x[y==1,1], s=15, c='blue', label ='Cluster_2')
plt.scatter(x[y==2,0], x[y==2,1], s=15, c='green', label='Cluster_3')
plt.scatter(x[y==3,0], x[y==3,1], s=15, c='cyan', label='Cluster_4')
plt.scatter(x[y==4,0], x[y==4,1], s=15, c='magenta', label='Cluster_5')

plt.scatter(kmeans5.cluster_centers_[:,0], kmeans5.cluster_centers_[:,1], s = 50, c='darkgreen', label='Centroids')
plt.legend()

plt.show()

# Narration:
# x[y==0,0] sepal length 0, 2nd 0 is fixed for 5 iteration ; x[y==0,1] sepal with 0,1 (1 is sepal width fixed for iteration)
# 1 to 4 is five iteration
# s = 15 is data points size/ depthness



```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#The output graph of the Elbow method is shown below. Note that the shape of elbow is approximately formed at k=3
#Plotting the results onto a line graph, allowing us to observe 'The elbow'

error=[]
for i in range(1, 11):
    kmeans11 = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init = 10, random_state=0).fit(x)
    kmeans11.fit(x)
    error.append(kmeans11.inertia_)
 
import matplotlib.pyplot as plt
plt.plot(range(1,11), error)
plt.title('Elbow using WCSS')
plt.xlabel('No of Cluster')
plt.ylabel('WCSS')
plt.show()
```

You can clearly see why it is called 'The elbow method' from the above graph, the optimum clusters is where the elbow occurs. This is when the within cluster sum of squares (WCSS) doesn't decrease significantly with every iteration. 


<h4 align="center"> where Yi is centroid for observation Xi.</h4>


Now that we have the optimum amount of clusters(3), we can move on to applying K-means clustering to the Iris dataset.

### Step 6: Now, let’s instantiate and fit our K means cluster model. We are going to use three clusters and a random state of 21.

```python
# As you can see, the optimal value of k is between 2 and 4, as the elbow-like shape is formed at k=3 in the above graph.
# Let’s implement k-means again using k=3

kmeans3 = KMeans(n_clusters=3, random_state=21)
y = kmeans3.fit_predict(x)
print(y)
```

### Step 7: With the following code you can identify the center points of the data.

```python
kmeans3.cluster_centers_
```

```python
#Applying kmeans to the dataset / Creating the kmeans classifier

plt.scatter(x[y==0,0], x[y==0,1], s = 15, c='red', label='Cluster1')
plt.scatter(x[y==1,0], x[y==1,1], s = 15, c='green', label='Cluster2')
plt.scatter(x[y==2,0], x[y==2,1], s = 15, c='blue', label='Cluster3')

plt.scatter(kmeans3.cluster_centers_[:,0], kmeans3.cluster_centers_[:,1], s = 35, c='black', label='Centroids')

plt.show()

# Overvations:
# Initially we have five clusters, now we can three clusters & its clustered properly
```

### Step 8: Now lets compare original data vs our cluster results using following code

```python
# plot idenfied clusters & compare with answere
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].scatter(x=df['sepal length (cm)'], y=df['sepal width (cm)'], c=y, cmap='gist_rainbow', edgecolor='k', s=150)
axes[1].scatter(x=df['sepal length (cm)'], y=df['sepal width (cm)'], c=iris.target, cmap='jet', edgecolor='k', s=150)
axes[0].scatter(kmeans3.cluster_centers_[:,0], kmeans3.cluster_centers_[:,1], s = 200, c='yellow', label='Centroids')
axes[1].scatter(kmeans3.cluster_centers_[:,0], kmeans3.cluster_centers_[:,1], s= 200, c='yellow', label='Centroids')

axes[0].set_xlabel('Petal Length', fontsize=18)
axes[0].set_ylabel('Patal Width', fontsize=18)
axes[1].set_xlabel('Petal Length', fontsize=18)
axes[1].set_ylabel('Petal Width', fontsize=18)

axes[0].tick_params(direction='in', length=10, width=5, colors='k', labelsize=10)  
axes[1].tick_params(direction='out', length=10, width=5, colors='k', labelsize=10) # direction='out'
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)
plt.show()

# Notes
# c-y: target (3 target setosa, versicolor, virginica)
# axes[0]: actual data/first plot, axes[1]: predicted /second plot
```

### Comparing the actual Petal Length & Petal Width with Prediction

```python
# plot idenfied clusters & compare with answere
fig, axes = plt.subplots(1,2, figsize=(14,6))

axes[0].scatter(x=df['petal length (cm)'], y=df['petal width (cm)'], c=y, cmap='gist_rainbow', edgecolor='k', s=150)
axes[1].scatter(x=df['petal length (cm)'], y=df['petal width (cm)'], c=iris.target, cmap='jet', edgecolor='k', s=150)
axes[0].set_xlabel('Petal Length', fontsize=18)
axes[0].set_ylabel('Petal Width', fontsize=18)
axes[1].set_xlabel('Petal Lenght', fontsize=18)
axes[1].set_ylabel('Petal Width', fontsize=18)
axes[0].tick_params(direction='in', length=10, width=5, color='k', labelsize=20)
axes[1].tick_params(direction='in', length=10, width=5, color='k', labelsize=20)
axes[0].set_title('Actual', fontsize=18)
axes[1].set_title('Predicted', fontsize=18)

plt.show()
```
