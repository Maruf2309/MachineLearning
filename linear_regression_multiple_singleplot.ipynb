<h3 style="color:purple" align="left" > Multiple Linear Regression in Single Plot </h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit # used to fit Line with slop & intercept
import random  # random module for generating random number
```

```python
# Define x, y data
x1 = np.linspace(1,100,20) # data range (1-100) total 20 datapoint
y1 = x1 + np.random.randint(10,50,20) # x + Noise
plt.scatter(x1, y1)
plt.show()
```

```python
# Lets Define multiple X, y
x1 = np.linspace(1,100,20) 
y1 = x1 + np.random.randint(10,50,20) # x + Noise

x2 = np.linspace(2,100,20)
y2 = -x2 + np.random.randint(10, 50, 20) # x + Noise

x3 = np.linspace(3, 100, 20)
y3 = 5*x1 + np.random.randint(10, 50, 20) # x + Noise

plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.scatter(x3, y3)


# Now for the line
# define information
def line(x,m,c):   # m:slop, c=Intercept
    return m*x+c

# for fitting all lines together used for loop
for y in (y1, y2, y3):
    for x in (x1,x2,x3):
        # fit line with optimizer parameter
        popt,cov = curve_fit(line, x,y)            # popt is an optimer params
    print(popt)                                    # popt is showing 3 different slop & Intercept
    

plt.show()
```
