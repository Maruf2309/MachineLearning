# Image Cropping

```python
image[100:1800,750:2400]  # Row & Col slicing
```

```python
io.imshow(image[100:1800,750:2400])
```

```python
image [:,:,2]  # Blue color channel, 2 means blue, RGB r-0; g-1, b-2
```

```python
image [:,: 2] > 180  # blue color channel > 180 pixel value
```

```python
blue_mask = image[:,:, 2] > 180   # numpy array all rows, all columns,  < will revers the image 
image [blue_mask] = 0 # 0 mean will make all pixel value zero
```

```python
io.imshow(image)
```

# Saving Image

```python
io.imsave('test.img', image)   # simpleitk required for image saving
```
